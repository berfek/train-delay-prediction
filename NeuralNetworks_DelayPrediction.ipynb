Cyclical Features Encoding with Sine and Cosine Transformation

```python
import pandas as pd
import numpy as np
```

```python
newdf['hour'] = newdf['BookedArrival.$date'].dt.hour
newdf['minute'] = newdf['BookedArrival.$date'].dt.minute
newdf['second'] = newdf['BookedArrival.$date'].dt.second
```

```python
#convert time to float
newdf['hourfloat']=newdf.hour+newdf.minute/60.0+ newdf.second/3600.0

```

```python
#sine and cosine transformation

newdf['sin_time']=np.sin(2.*np.pi*newdf.hourfloat/24.)
newdf['cos_time']=np.cos(2.*np.pi*newdf.hourfloat/24.)

newdf
```

```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from sklearn.utils.class_weight import compute_class_weight

cols = ['sin_time', 'cos_time', 'day_of_week', 'Cluster', 'Longitude', 'Latitude']
X = df2[cols]

y = df2['DelayCategory']
y = y.to_frame()

# Convert 'DelayCategory' column to integers if it's not already
y['DelayCategory'] = y['DelayCategory'].astype(int)

# Calculate class weights
class_weights = compute_class_weight('balanced', classes=y['DelayCategory'].unique(), y=y['DelayCategory'])

# Create a dictionary with class labels as keys and corresponding weights as values
class_weights_dict = dict(zip(y['DelayCategory'].unique(), class_weights))

print("Class Weights:")
print(class_weights_dict)

#Apply oversampling
oversampler = RandomOverSampler(random_state=42)
X_resampled, y_resampled = oversampler.fit_resample(X, y)

#Convert the oversampled arrays back to DataFrame
X_resampled = pd.DataFrame(X_resampled, columns=cols)
y_resampled = pd.DataFrame(y_resampled, columns=['DelayCategory'])

#XCheck the class distribution after oversampling
print("Class Distribution after Oversampling:")
print(y_resampled['DelayCategory'].value_counts())
```

Train NN with oversampled data

```python
train_dataset, test_dataset, train_labels, test_labels = train_test_split(
    X_resampled, y_resampled, test_size=0.2, stratify=y_resampled, random_state=42
)
```

```python

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping

model = Sequential()
model.add(Dense(128, activation='relu', input_dim=train_dataset.shape[1]))
model.add(Dense(4, activation='softmax'))
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
early_stopping = EarlyStopping(monitor='val_loss', patience=5, verbose=1, restore_best_weights=True)
hist = model.fit(train_dataset, train_labels, epochs=50, batch_size=16, validation_split=0.2, callbacks=[early_stopping])

```
